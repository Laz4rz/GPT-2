### How variance grows in residual streams

```python
import matplotlib.pyplot as plt
import math
import numpy as np

x = torch.randn(768)
temporary_x_std = [x.std()]
temporary_x_std_normed = [x.std()]
for i in range(100):
    x += torch.randn(768)
    temporary_x_std.append(x.std())
    temporary_x_std_normed.append(x.std()/math.sqrt(i+2))

plt.figure(figsize=(10, 3))
plt.title("Scaled vs non-scaled std of stacked layers in residual streams")
plt.plot(temporary_x_std, marker="o", alpha=0.5, label="non-scaled")
plt.plot(temporary_x_std_normed, marker="o", alpha=0.5, label="scaled")
plt.yticks(np.arange(0, 10, 1))
plt.grid()
plt.legend()
```

```python
import matplotlib.pyplot as plt
import math
import numpy as np

x = torch.randn(768)
temporary_x_std = [x.std()]
temporary_x_std_normed = [x.std()/100]
for i in range(100):
    x += torch.randn(768)
    temporary_x_std.append(x.std())
    temporary_x_std_normed.append(x.std()/math.sqrt(100))

plt.figure(figsize=(10, 3))
plt.title("Scaled vs non-scaled std of stacked layers in residual streams")
plt.plot(temporary_x_std, marker="o", alpha=0.5, label="non-scaled")
plt.plot(temporary_x_std_normed, marker="o", alpha=0.5, label="scaled")
plt.yticks(np.arange(0, 10, 1))
plt.ylabel("std")
plt.xlabel("n_layer")
plt.grid()
plt.legend()
```

```python
import torch
import numpy as np
from torch.nn import functional as F
from gpt2 import softmax, safe_softmax, online_safe_softmax, online_safe_softmax_list

torch.manual_seed(42)

# if you dont cast long to float F.softmax throws 
# RuntimeError: "softmax_lastdim_kernel_impl" not implemented for 'Long' lol
x = torch.tensor([1000, 999, 998], dtype=torch.float)
x = torch.randn(5000)
s = softmax()
ss = safe_softmax()
oss = online_safe_softmax()
ossl = online_safe_softmax_list()

ground_truth = F.softmax(x, dim=0)
rs = s(x, np)
rss = ss(x, np)
ross = oss(x, np)
rossl = ossl(x, np)

print(ground_truth)
print(rs)
print(rss)
print(ross)

print(f"Ground truth and custom softmax are equal: {torch.allclose(ground_truth, rs)}")
print(f"Ground truth and safe softmax are equal: {torch.allclose(ground_truth, rss)}")
print(f"Ground truth and online safe softmax are equal: {torch.allclose(ground_truth, ross)}")
print(f"Ground truth and online safe softmax list are equal: {torch.allclose(ground_truth, rossl)}")
```

```python
%timeit s(x, np)
%timeit ss(x, np)
%timeit oss(x, np)
%timeit ossl(x, np)
%timeit F.softmax(x, dim=0)
```

```python
import timeit
results = {
    s: {"mean": [], "std": []},
    ss: {"mean": [], "std": []},
    oss: {"mean": [], "std": []},
    ossl: {"mean": [], "std": []},
    F.softmax: {"mean": [], "std": []}
}
for n in [10, 100, 1000, 10000]:
    x = torch.randn(n)
    results[s]["mean"].append(timeit.timeit(lambda: s(x, np), number=10))
    results[ss]["mean"].append(timeit.timeit(lambda: ss(x, np), number=10))
    results[oss]["mean"].append(timeit.timeit(lambda: oss(x, np), number=10))
    results[ossl]["mean"].append(timeit.timeit(lambda: ossl(x, np), number=10))
    results[F.softmax]["mean"].append(timeit.timeit(lambda: F.softmax(x, dim=0), number=10))
        
```
